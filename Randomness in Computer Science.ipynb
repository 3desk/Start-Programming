##  Randomness in Computer Science.
Concept:  random(), random.randint(), random.random()

**Random:** means unexpected. In computer science, it means computer makes a choice that is unexpected or unpredictable. 

**Deterministic:** means predetermined, predictable. Computers are deterministic i.e. their actions are predetermined by the instructions given to them. The output of the instructions given to a computer is predictable and if someone makes an error while giving instructions, an error is predictable. Again, computers are **deterministic**.

**Random module:** is used to make a deterministic machine like computer make **random decisions/choices**. Random module has functions that generate random numbers using pseudo random number generators.

```python
# Following statement imports  the random module.
import random

# Following statement uses randint() function of random module to generate a random number between 1 & 5.
# The random number generated by the function is then pointed by variable random_integer.
random_integer = random.randint(1,5)

# Following statement prints the random number that is generated, so that we may see it.
print(random_integer)

# Following function generates decimal numbers between 0.0 to 1.0 (excluding 1.0).
random_float = random.random()

print(random_float)
```
